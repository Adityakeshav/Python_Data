```python
# Load necessary library
library(e1071)

# Sample data: a data frame with text and labels
data <- data.frame(
    text = c("Free money", "Call now", "Important update", "Meeting at 10", "Win a prize", "Project deadline"),
    label = c("spam", "spam", "ham", "ham", "spam", "ham")
)

# Create a Document-Term Matrix
library(tm)
corpus <- Corpus(VectorSource(data$text))
dtm <- DocumentTermMatrix(corpus, control = list(wordLengths = c(1, Inf)))

# Convert the Document-Term Matrix to a data frame
dtm_df <- as.data.frame(as.matrix(dtm))
dtm_df$label <- data$label

# Split the data into training and testing sets
set.seed(123)
train_indices <- sample(1:nrow(dtm_df), 0.7 * nrow(dtm_df))
train_data <- dtm_df[train_indices, ]
test_data <- dtm_df[-train_indices, ]

# Train the Naive Bayes model
model <- naiveBayes(label ~ ., data = train_data)

# Predict on the test set
predictions <- predict(model, test_data)

# Evaluate the model
confusion_matrix <- table(predictions, test_data$label)
print(confusion_matrix)
```

```python
# Load necessary libraries
library(e1071)
library(tm)

# Sample data: a data frame with text and labels
data <- data.frame(
    text = c("Free money", "Call now", "Important update", "Meeting at 10", "Win a prize", "Project deadline"),
    label = c("spam", "spam", "ham", "ham", "spam", "ham")
)

# Create a Corpus
corpus <- Corpus(VectorSource(data$text))

# Inspect the Corpus
inspect(corpus)

# Clean the text data
corpus <- tm_map(corpus, content_transformer(tolower)) # Convert to lowercase
corpus <- tm_map(corpus, removePunctuation) # Remove punctuation
corpus <- tm_map(corpus, removeNumbers) # Remove numbers
corpus <- tm_map(corpus, removeWords, stopwords("english")) # Remove stopwords
corpus <- tm_map(corpus, stripWhitespace) # Strip whitespace

# Inspect the cleaned Corpus
inspect(corpus)

# Create a Document-Term Matrix
dtm <- DocumentTermMatrix(corpus, control = list(wordLengths = c(1, Inf)))

# Convert the Document-Term Matrix to a data frame
dtm_df <- as.data.frame(as.matrix(dtm))
dtm_df$label <- data$label

# Split the data into training and testing sets
set.seed(123)
train_indices <- sample(1:nrow(dtm_df), 0.7 * nrow(dtm_df))
train_data <- dtm_df[train_indices, ]
test_data <- dtm_df[-train_indices, ]

# Train the Naive Bayes model
model <- naiveBayes(label ~ ., data = train_data)

# Predict on the test set
predictions <- predict(model, test_data)

# Evaluate the model
confusion_matrix <- table(predictions, test_data$label)
print(confusion_matrix)
```
